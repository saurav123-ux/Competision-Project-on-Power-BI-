```python
col=['age','watched_percentage']
```

```python
for i in col:
    p25=df[i].quantile(0.25)
    p75=df[i].quantile(0.75)
    iqr=p75-p25
    ll=p25-1.5*(iqr)
    ul=p75+1.5*(iqr)
    df=df[df[i]<=ul]
    df=df[df[i]>=ll]
    plt.figure()
    sns.boxplot(y=i,data=df)
```
